# Manipulation-2

```python
import pandas as pd
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
column_seats=pd.Series([5,5,5,5,5])
car_sales["Seats"] = column_seats
car_sales
```

```python
car_sales.fillna(5,inplace=True)
```

```python

car_sales
```

```python
#creating fuel used column
fuel_economy=[7.2,8.2,5.5,6.2,7.2,8.3,9.4,6.7,7.2,8.5]
car_sales["fuel_used"]=fuel_economy
car_sales
```

```python
car_sales["fuel_used_per_100KM"]=fuel_economy
car_sales
```

```python
#deleting a column
car_sales.drop(["fuel_used"],axis=1,inplace=True)
```

```python
car_sales.drop(["fuel_used_per_100KM"],axis=1,inplace=True)
car_sales
```

```python
#creating a column
car_sales["total fuel used(L)"] = car_sales["Odometer (KM)"/100 * car_sales["fuel_used(L)"]
car_sales
```
